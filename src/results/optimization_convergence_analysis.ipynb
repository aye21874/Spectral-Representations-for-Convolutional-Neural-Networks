```python
from matplotlib import pyplot as plt
import pickle


history_files=["3x3_generic_spatial","3x3_generic_spectral",
               "5x5_generic_spatial","5x5_generic_spectral",
                "3x3_spatial","3x3_deep_spectral",
               "5x5_deep_spatial","5x5_deep_spectral"]


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

for i in range(1, 5):
    plt.subplot(2, 2, i)
    
    # Loading training histories

    file = open(history_files[2*(i-1)],'rb')
    history_spatial = pickle.load(file)
    file.close()

    file = open(history_files[2*i-1],'rb')
    history_spectral = pickle.load(file)
    file.close()

    error_map={}
    error_map["spatial"]=[]
    error_map["spectral"]=[]

    for i in range(len(history_spatial["accuracy"])):
        error_map["spatial"].append(1-history_spatial["accuracy"][i])
        error_map["spectral"].append(1-history_spectral["accuracy"][i])

        plt.plot(error_map["spatial"], color="lightblue")
        plt.plot(error_map["spectral"], color="orange")
        plt.ylabel('Training Error')
        plt.xlabel('Number of epochs')

cols=["Deep","Generic"]
rows=["3x3","5x5"]

plt.legend(['spatial', 'spectral'], loc='upper right')

fig.text(0.30, 0.9, 'Generic', ha='center',fontsize=15)
fig.text(0.72, 0.9, 'Deep', ha='center',fontsize=15)

fig.text(0.05, 0.70, '3x3', va='center', rotation='horizontal',fontsize=15)
fig.text(0.05, 0.30, '5x5', va='center', rotation='horizontal',fontsize=15)

plt.savefig("error_analysis.png")
plt.show()
```
